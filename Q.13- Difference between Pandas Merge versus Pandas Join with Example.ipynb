# Explaining with an example the difference between Pandas.merge() vs Pandas.join()

```python
# importing library

import pandas as pd
import numpy as np
```

# MERGE

--> It combines data on basis of common columns                               
-->Search for same columns name then look for their row values that have the same columns name        
-->If above thing matches then it merges, basically it clubs columns together and form one dataframe                          
-->By default it does Inner join only

```python
# Creating two Example DataFrames 

df1=pd.DataFrame({"ID":[1,2,3,4],
                 "C1":[1,2,3,4],
                 "C2":[6,7,8,9],
                 "C3":[10,11,12,13]})
df2=pd.DataFrame({"ID":[1,1,3,4],
                 "CA":[8,9,10,11],
                 "CB":[5,3,2,1],
                "C3":[10,12,12,13]})
```

```python
# Visualize df1

df1
```

```python
# Visualize df1

df2
```

```python
merge= pd.merge(df1,df2)
merge
```

```python
pd.merge(df1,df2,on="ID")
```

# JOINS

-->Inner: same as merge process                                                                               
-->Outer: Joins all rows from both DataFrame, no data is lost                                                
-->left: left table is completely included and from right only those included who's value matches with left                  
-->Right:Right table is completely included and from left only those included who's value matches with Right

```python
df1
```

```python
df2
```

```python
#outer
pd.merge(df1,df2,on="C3",how="outer")
```

```python
#left 
pd.merge(df1,df2,how="left")
```
